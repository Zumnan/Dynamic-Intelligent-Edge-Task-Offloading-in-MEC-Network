```python
import os
import pandas as pd

# Define the path to the dataset folder
dataset_path = r'C:\Users\Administrator\Downloads\Edge-Computing-Dataset-master\Edge-Computing-Dataset-master\Data'

# Define the relevant features for task offloading evaluation
selected_features = ['Cell Identity(CI)', 'User Agent', 'Downlink traffic', 'Location Area Code(LAC)', 'Duration',
                     'Start Time', 'Radio Access Type(RAT)', 'End Time', 'Uplink traffic', 'Destination IP',
                     'URL', 'Source IP']

# Iterate over the subfolders in the dataset path (shanghai and guangzhou)
for subfolder in ['shanghai', 'guangzhou']:
    subfolder_path = os.path.join(dataset_path, subfolder)

    # Iterate over the files in each subfolder
    for file_name in os.listdir(subfolder_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(subfolder_path, file_name)

            # Read the CSV file and select the relevant features
            df = pd.read_csv(file_path)
            df = df[selected_features]

            # Convert 'Start Time' and 'End Time' columns to datetime format
            df['Start Time'] = pd.to_datetime(df['Start Time'])
            df['End Time'] = pd.to_datetime(df['End Time'])

            # Function to determine the offloading decision based on available information
            def determine_offloading_decision(row):
                # Example decision logic
                if row['Downlink traffic'] > row['Uplink traffic']:
                    return 'Edge Cloud'
                else:
                    return 'Mobile Device'

            # Apply the determine_offloading_decision function to each row in the dataset
            df['Offloading Decision'] = df.apply(determine_offloading_decision, axis=1)

            # Function to calculate latency for each task based on the offloading decision
            def calculate_latency(row):
                edge_cloud_processing_time = 5.0  # Example value for edge cloud processing time (you can change this)
                if row['Offloading Decision'] == 'Edge Cloud':
                    return (row['End Time'] - row['Start Time']).total_seconds() + edge_cloud_processing_time
                else:
                    return row['Duration']

            # Apply the calculate_latency function to each row in the dataset
            df['Latency'] = df.apply(calculate_latency, axis=1)

            # Calculate throughput for each task
            df['Throughput'] = (df['Uplink traffic'] + df['Downlink traffic']) / df['Latency']

            # Calculate resource utilization percentage
            edge_cloud_utilization_percentage = (df[df['Offloading Decision'] == 'Edge Cloud']['Latency'].sum() / df['Latency'].sum()) * 100

            # Save the updated dataset back to the original CSV file
            df.to_csv(file_path, index=False)

            # Print the first few rows of the updated dataset for verification
            print(df.head())

            # Print the evaluation metrics
            print("Edge Cloud Utilization Percentage:", edge_cloud_utilization_percentage)
```

```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.ensemble import VotingClassifier
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model

# Define the path to the dataset folder
dataset_path = r'C:\Users\Administrator\Downloads\Edge-Computing-Dataset-master\Edge-Computing-Dataset-master\Data'

# Define the relevant features for task offloading evaluation
selected_features = ['Cell Identity(CI)', 'User Agent', 'Downlink traffic', 'Location Area Code(LAC)', 'Duration',
                     'Start Time', 'Radio Access Type(RAT)', 'End Time', 'Uplink traffic', 'Destination IP',
                     'URL', 'Source IP', 'Offloading Decision', 'Latency']

# Create an empty DataFrame to store the combined dataset
combined_dataset = pd.DataFrame()

# Iterate over the subfolders in the dataset path (shanghai and guangzhou)
for subfolder in ['shanghai', 'guangzhou']:
    subfolder_path = os.path.join(dataset_path, subfolder)
    
    # Iterate over the files in each subfolder
    for file_name in os.listdir(subfolder_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(subfolder_path, file_name)
            
            # Read the CSV file and select the relevant features
            df = pd.read_csv(file_path)
            df = df[selected_features]
            
            # Append the DataFrame to the combined dataset
            combined_dataset = combined_dataset.append(df, ignore_index=True)

# Prepare the dataset for the reinforcement learning agent
X = combined_dataset[['Downlink traffic', 'Duration', 'Latency']].values
y = combined_dataset['Offloading Decision'].values

# Convert categorical labels to numeric values for RL training
unique_labels = combined_dataset['Offloading Decision'].unique()
label_to_index = {label: index for index, label in enumerate(unique_labels)}
index_to_label = {index: label for label, index in label_to_index.items()}
y_numeric = np.array([label_to_index[label] for label in y])

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_numeric, test_size=0.2, random_state=42)

# Scale the features for better model performance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Implement the Deep Q-Network (DQN) model
# ...

# Train the DQN model
# ...

# Save the trained DQN model
# ...

# Load the trained DQN model
model = load_model('dqn_model.h5')

# Function to make offloading decisions using the trained DQN model
def make_offloading_decision_dqn(traffic_load, duration, latency):
    input_data = scaler.transform([[traffic_load, duration, latency]])
    decision_index = np.argmax(model.predict(input_data)[0])
    decision = index_to_label[decision_index]
    return decision

# Define function to simulate MEC network environment and conduct experiments
def simulate_mec_network():
    # Define base station locations, user access time, and other relevant network information
    base_stations = combined_dataset[['Location Area Code(LAC)', 'Cell Identity(CI)']].drop_duplicates().values
    user_access_time = combined_dataset[['Start Time', 'End Time']].values

    # Simulate experiments on a representative subset of the dataset
    subset_size = 1000  # Define the size of the subset
    subset_dataset = combined_dataset.sample(n=subset_size, random_state=42)

    # Apply the decision-making logic and load balancing to each row in the subset dataset
    subset_dataset['Offloading Decision DQN'] = subset_dataset.apply(lambda row: make_offloading_decision_dqn(
        row['Downlink traffic'], row['Duration'], row['Latency']), axis=1)

    # Calculate performance metrics for the subset dataset
    actual_decisions = subset_dataset['Offloading Decision']
    predicted_decisions = subset_dataset['Offloading Decision DQN']

    # Calculate precision, recall, and F1-score
    report = classification_report(actual_decisions, predicted_decisions)
    print("Classification Report (Subset Dataset):")
    print(report)

    # Calculate confusion matrix
    cm = confusion_matrix(actual_decisions, predicted_decisions)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix (Subset Dataset)')
    plt.show()

# Call the function to simulate the MEC network and conduct experiments
simulate_mec_network()

# Create and train alternative models

# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train_scaled, y_train)
predicted_labels_rf = rf_model.predict(X_test_scaled)
report_rf = classification_report(y_test, predicted_labels_rf)
print("Random Forest Classification Report:")
print(report_rf)

# Gradient Boosting
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train_scaled, y_train)
predicted_labels_gb = gb_model.predict(X_test_scaled)
report_gb = classification_report(y_test, predicted_labels_gb)
print("Gradient Boosting Classification Report:")
print(report_gb)

# Support Vector Machines (SVM)
svm_model = SVC()
svm_model.fit(X_train_scaled, y_train)
predicted_labels_svm = svm_model.predict(X_test_scaled)
report_svm = classification_report(y_test, predicted_labels_svm)
print("SVM Classification Report:")
print(report_svm)

# Ensemble Model (Voting Classifier)
ensemble_model = VotingClassifier(estimators=[('rf', rf_model), ('gb', gb_model), ('svm', svm_model)])
ensemble_model.fit(X_train_scaled, y_train)
predicted_labels_ensemble = ensemble_model.predict(X_test_scaled)
report_ensemble = classification_report(y_test, predicted_labels_ensemble)
print("Ensemble Model Classification Report:")
print(report_ensemble)
```
